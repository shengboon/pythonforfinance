```python
import numpy as np
import scipy.stats as sp
```

```python
def eucall(t, s, k, r, q, v):
    if t <= 0:
        return max(s - k, 0)
    else:
        d1 = (1/(v*np.sqrt(t))) * (np.log(s/k)+ (r-q+0.5*v**2)*t)
        d2 = d1 - v*np.sqrt(t)
        return s*np.exp(-q*t)*sp.norm().cdf(d1)-sp.norm().cdf(d2)*k*np.exp(-r*t)
```

```python
def euput(t, s, k, r, q, v):
    if t <= 0:
        return max(k-s, 0)
    else:
        d1 = (1/(v*np.sqrt(t))) * (np.log(s/k)+ (r-q+0.5*v**2)*t)
        d2 = d1 - v*np.sqrt(t)
        return -s*np.exp(-q*t)*sp.norm().cdf(-d1)+sp.norm().cdf(-d2)*k*np.exp(-r*t)
```

```python
s = np.arange(90,140,1)
y = eucall(2,s,120,0.03,0.02,0.2)
yp=euput(2,s,120,0.03,0.02,0.2)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
fig ,ax= plt.subplots(1,2,figsize=(15,8))
ax[0].plot(s,y,label='t=2')
y = eucall(0.00001,s,120,0.03,0.02,0.2)
ax[0].plot(s,y,label='Maturity ($t=0$)')
ax[0].legend(frameon=False)
ax[1].plot(s,yp,label='t=2')
y = euput(0.00001,s,120,0.03,0.02,0.2)
ax[1].plot(s,y,label='Maturity ($t=0$)');
ax[0].set_title('European Call')
ax[1].set_title('European Put')
```

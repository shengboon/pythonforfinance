```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
sns.set()
```

```python
import numpy as np
```

```python
nbr_points = 15
a = np.random.randn(nbr_points)
x = np.arange(0,nbr_points)
y = x + a
```

```python
fig , ax = plt.subplots(1,1,figsize=(15,8))
ax.scatter(x,y,s=100);
ax.plot(x,y,'r');
ax.set_xlabel('X')
ax.set_ylabel('Y');
```
